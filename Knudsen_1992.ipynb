```python
import numpy as np
import matplotlib.pyplot as plt

# Define constants
mu_0 = 4 * np.pi * 1e-7  # Vacuum permeability (H/m)
V_A = 1e6  # Alfvén speed (m/s), example value
Gamma = 0.5  # Reflection coefficient, example value
z = 1e3  # Propagation distance (m), example value

# Define frequency range
omega = np.linspace(1, 1e6, 1000)  # Frequency range (rad/s)

# Calculate the function and its modulus
x = np.exp(-2j * omega * z / V_A)
Z_hat = mu_0 * V_A * (1 + Gamma * x) / (1 - Gamma * x)
Z_hat_modulus = np.abs(Z_hat)

# Plot the real and imaginary parts of Z_hat
plt.figure(figsize=(10, 5))
plt.plot(omega, Z_hat.real, label="Real part of Z_hat(ω)", linestyle='-', color='blue')
plt.plot(omega, Z_hat.imag, label="Imaginary part of Z_hat(ω)", linestyle='--', color='red')
plt.xlabel("Frequency (ω, rad/s)")
plt.ylabel("Z_hat(ω)")
plt.title("Real and Imaginary Parts of Z_hat(ω)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot the modulus of Z_hat
plt.figure(figsize=(10, 5))
plt.plot(omega, Z_hat_modulus, label="|Z_hat(ω)|", color='green')
plt.xlabel("Frequency (ω, rad/s)")
plt.ylabel("|Z_hat(ω)|")
plt.title("Modulus of Z_hat(ω)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
