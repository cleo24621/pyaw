# rotation matrix: NEC to VFM

```python
def get_rotation_matrix(VsatN,VsatE):
    theta = np.arctan(VsatE/VsatN)

    cos_theta = np.cos(theta)
    sin_theta = np.sin(theta)

    # Stack components to construct the rotation matrices
    rotation_matrix = np.array([
        [cos_theta, sin_theta],
        [-sin_theta, cos_theta]
    ])

    # Transpose axes to create a (3, 2, 2) array
    rotation_matrix = np.transpose(rotation_matrix, (2, 0, 1))

    return rotation_matrix

def do_rotation(vectors_N, vectors_E,rotation_matrix):
    vectors_NE = np.stack((vectors_N, vectors_E), axis=1)
    return np.einsum('nij,nj->ni', rotation_matrix, vectors_NE)

do_rotation(VsatN,VsatE,get_rotation_matrix(VsatN,VsatE))
```

```python
# Compute norms (lengths) before and after rotation
original_norms = np.linalg.norm(Vsat_vectors_NE, axis=1)
rotated_norms = np.linalg.norm(rotated_vectors, axis=1)

# Check if norms are close
print("Original norms:", original_norms)
print("Rotated norms:", rotated_norms)
print("Are norms preserved?", np.allclose(original_norms, rotated_norms))
```
